```python
from pathlib import Path
import pandas as pd

from tqdm.notebook import tqdm
```

```python
table_names = [
    "IND",
    "DRUG",
    "ADR",
    "DEMO",
    "OUT",
    "SRCE",
    "FOLLOWUP",
    "LINK",
]
```

```python
# table_name = "DRUG"
# batch = "20210901"  # OLD
batch = "20220601" # NEW
parquet_path = Path.cwd().parent / "data" / "to_parquet" / batch
partitioned_path = parquet_path / "partitioned"
fpath_input = partitioned_path
fpath_output = parquet_path
```

```python
for table_name in tqdm(table_names):
    fname = f"{table_name}.parquet"
    fpath = fpath_output / fname
    if fpath.exists():
        continue
    
    results = []
    files = list(fpath_input.glob(f"**/{table_name}*.parquet"))
    for file in tqdm(files):
        results.append(pd.read_parquet(file))
    
    result = pd.concat(results, axis="rows", ignore_index=True)
    result.to_parquet(fpath)
```
